```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Example data: engine size vs mileage
engine_size = np.array([1.0, 1.3, 1.5, 2.0, 2.2, 2.5, 3.0]).reshape(-1, 1)  # in liters
mileage = np.array([22, 20, 18, 15, 14, 13, 11])  # in mpg

# Create the model
model = LinearRegression()

# Train the model
model.fit(engine_size, mileage)

# Coefficients
b1 = model.coef_[0]  # Slope
b0 = model.intercept_  # Intercept

# Print coefficients
print(f"Coefficient (b1): {b1}")
print(f"Intercept (b0): {b0}")

# Predicted mileage
predicted_mileage = model.predict(engine_size)

# Plotting the regression line
plt.scatter(engine_size, mileage, color="blue", label="Original data")
plt.plot(engine_size, predicted_mileage, color="red", label="Fitted line")
plt.xlabel("Engine Size (Liters)")
plt.ylabel("Mileage (MPG)")
plt.legend()
plt.show()

# Predicting mileage for a new engine size
new_engine_size = np.array([[1.8]])
predicted_value = model.predict(new_engine_size)
print(f"Predicted mileage for engine size {new_engine_size[0][0]}L: {predicted_value[0]:.2f} MPG")
```
